## iris 데이터

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
X
```

### PCA
- 주성분 분석을 사용해 4차원 데이터를 2차원으로 축소

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components =2) # 전체 데이터를 2차원으로 축소
```

```python
pca.fit(X)
```

```python
X_trans = pca.transform(X)
X_trans
```

```python
# 축소된 데이터 시각화
plt.figure(figsize=(10,10))
# 산점도
plt.scatter(X_trans[:,0], # 축소된 0번 컬럼
            X_trans[:,1], # 축소된 1번 컬럼
            c = y) 
plt.show()
```

### 사람 데이터

```python
from sklearn.datasets import fetch_lfw_people
```

```python
# 125*94(세로 * 가로)
# 87 *65
people = fetch_lfw_people(min_faces_per_person = 20, resize = 0.7)
```

```python
people.keys()
```

```python
people.data.shape
```

```python
people.target
```

```python
people.target_names
```

```python
plt.imshow(people.images[516], cmap = 'gray')
```

```python
X = people['data']
y = people['target']
```

```python
#픽셀값 = 0~255
X[0]
```

```python
# 픽셀값 범위를 0~1
X = X / 255.
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,
                                                   random_state = 10)
```

```python
# knn모델 학습
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
knn.score(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
pca = PCA(n_components = 100, whiten = True)
```

```python
pca.fit(X_train)
```

```python
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
knn.fit(X_train_pca, y_train)
```

```python
knn.score(X_train_pca, y_train)
```

```python
knn.score(X_test_pca, y_test)
```

```python
# 주성분 데이터 확인
pca.components_[0].shape
```

```python
plt.imshow(pca.components_[1].reshape(87,65))
plt.show()
```

```python
# 주성분으로 이미지 확인하기
mglearn.plots.plot_pca_faces(X_train, X_test , (87,65))
```
